#Add New columns

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as pc
pio.templates.default = "plotly_white"
```

```python
gp=df.groupby("Order Month").agg({"Sales":"sum"}).reset_index()
# plt.grid(True)
# # sns.barplot(x=gp.index,y=gp.Sales,palette="rocket")
# plt.plot(gp.index, gp.Sales, color='Red', marker='o', linestyle='dotted', label='Sales Trend')
# plt.title("Monthly Sales")

# plt.show()
fig = px.line(gp,
              x='Order Month',
              y='Sales',
              title='Monthly Sales Analysis')
fig.show()
```

#conclusion: higest sales in 11 month
#conclusion:lowest sales in 2 month

```python
gp1=df.groupby("Category").agg({"Sales":"sum"}).reset_index()

fig=px.pie(gp1,values="Sales",names="Category",title="Category vs Sales",hole=0.2)
fig.show()
# sns.barplot(x=gp1.index,y=gp1.Sales,palette="rocket")
# plt.title("Category vs Sales")

# plt.show()
```

#conclusion: higest category is a Technology
#conclusion:lowest Category is a Furniture

```python
gp2=df.groupby("Sub-Category").agg({"Sales":"sum"}).reset_index()

fig = px.bar(gp2, x="Sub-Category", y="Sales", title="Sales Vs Sub-Category")
fig.show()


# sns.barplot(x=gp2.index,y=gp2.Sales,palette="rocket

# sns.barplot(x=gp2.index,y=gp2.Sales,palette="deep")
# plt.title("Sub-Category vs Sales")
#

# plt.show()
```

```python
gp3=df.groupby("Order Month").agg({"Profit":"sum"}).reset_index()

fig = px.line(gp3,
              x='Order Month',
              y='Profit',
              title='Monthly Profit Analysis')
fig.show()
# sns.barplot(x=gp3.index,y=gp3.Profit,palette="deep")
# plt.title("Total Profit vs Month")


# plt.show()
```

#Conclusion:Highest Profit in December(12)
#Conclusion:Lowest Profit in February(2)


```python
gp4=df.groupby("Category").agg({"Profit":"sum"}).reset_index()

fig=px.pie(gp4,values="Profit",names="Category",title="Category vs profit",hole=0.2)
fig.show()
```

```python
gp5=df.groupby("Sub-Category").agg({"Profit":"sum"}).reset_index()

fig=px.pie(gp5,values="Profit",names="Sub-Category",title="Category vs profit",hole=0.2)
fig.show()
```

```python
gp6=df.groupby("Segment").agg({"Sales":"sum","Profit":"sum"}).reset_index()
segments = gp6["Segment"]
sales = gp6["Sales"]
profit = gp6["Profit"]


fig = go.Figure()


fig.add_trace(go.Bar(
    x=segments,
    y=sales,
    name="Sales",
    marker_color="blue"
))


fig.add_trace(go.Bar(
    x=segments,
    y=profit,
    name="Profit",
    marker_color="orange"
))


fig.update_layout(
    barmode="group",  # Grouped bars
    title=" Sales vs Profit by Segment",
    xaxis_title="Segment",
    yaxis_title="Amount",
    legend_title="Metrics",
    template="plotly_white"  # Optional: For a clean look
)


fig.show()
```
